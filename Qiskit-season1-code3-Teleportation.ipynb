```python
from qiskit import *
import matplotlib
from qiskit.tools.visualization import plot_histogram, plot_bloch_multivector
```

```python
#In quantum computers the act of trying to transfer by copying isn't allowed
#becaouse the moment you copy, you're actually doing a measurement which 
#destorys the quantum state (information is lost)
```

```python
# Take advantage of entanglement
circuit = QuantumCircuit(3,3)
circuit.draw(output='mpl')
```

```python
#teleporting the state q0 to q2
circuit.x(0)
circuit.draw(output='mpl')
```

```python
circuit.barrier()
circuit.h(1)
circuit.cx(1,2)
```

```python
circuit.draw(output='mpl')
```

```python
circuit.barrier()
circuit.cx(0,1)
circuit.h(0)
circuit.draw(output='mpl')
```

```python
circuit.barrier()
circuit.measure([0,1],[0,1])
```

```python
circuit.draw(output='mpl')
```

```python
circuit.barrier()
circuit.cx(1,2)
circuit.cz(0,2)
circuit.draw(output='mpl')
```

```python
circuit.measure(2,2)
simulator = Aer.get_backend('qasm_simulator')
result = execute(circuit, backend = simulator, shots =1024).result()
counts = result.get_counts()
plot_histogram(counts)
```

```python
# In qiskit it is q2q1q0 as a result so q2 is always 1 and we transformed
# q0 = 0 to q0 = 1 in the first X gate part. Now it is copied or teleport
# to the q2 
# now we try to teleport the |+> state ((|0>+|1>)/sqrt(2)) 
```
